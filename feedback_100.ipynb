## Exercise 1

```python
import numpy as np

def calculate_pi(num):
    N_in = 0
    for _ in range(num):
        x = np.random.uniform(-1, 1)
        y = np.random.uniform(-1, 1)
        if x**2 + y**2 < 1:
            N_in += 1

    p = 4 * N_in / num
    return p

result = calculate_pi(1000)
print(result)
```

## Exercise 2

```python
import matplotlib.pyplot as plt
import numpy as np

def calculate_pi(num):
    N_in = 0
    for i in range(num):
        x1= np.random.uniform(-1, 1)
        y1 = np.random.uniform(-1, 1)
        if x1**2 + y1**2 < 1:
            N_in += 1

    p = 4 * N_in / num
    return p

pi_values_1 = [calculate_pi(10**2) for i in range(10000)]
pi_values_2 = [calculate_pi(10**3) for i in range(1000)]

plt.hist(pi_values_2, alpha=0.5, density=True, label="N=1000")

plt.hist(pi_values_1, alpha=0.5, density=True, label="N=100")

plt.legend()

plt.xlabel('estimated')
plt.ylabel('frequency')
plt.show()
```

## Exercise 3

```python

import numpy as np
import matplotlib.pyplot as plt

num = 10**6
x = np.random.uniform(-1, +1, num)
y = np.random.uniform(-1, +1, num)
Nin = 0
pi_list = []

for i in range(num):
    r = x[i]**2 + y[i]**2
    if r < 1:
        Nin = Nin + 1
    pi_values = 4 * Nin / (i + 1)
    pi_list.append(pi_values)

pi_values = np.array(pi_list)

difference_magnitude = np.abs(pi_values - np.pi)

plt.figure(figsize=(10, 6))
plt.plot(np.arange(1, num + 1), difference_magnitude, label='| - _true|')


x_plot = np.linspace(1, num, num=num)
y_plot = x_plot**(-0.5)
plt.plot(x_plot, y_plot, label='1/n', linestyle='--', color='red')


plt.plot(difference_magnitude)

plt.xscale('log')
plt.yscale('log')

plt.xlabel('Number')
plt.ylabel('Uncertainty')
plt.show()
```
